```python
import cv2
import os
from pathlib import Path

def extract_frames(video_path, output_base_folder, interval=225):
    """
    Extract every `interval`-th frame from a video and save them to a subfolder named after the video.
    """
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_folder = os.path.join(output_base_folder, video_name)
    os.makedirs(output_folder, exist_ok=True)

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"❌ Error: Could not open video: {video_path}")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"🎥 Processing {video_name} — {total_frames} frames at {fps:.2f} FPS")

    frame_count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % interval == 0:
            filename = os.path.join(output_folder, f"{video_name}_frame_{saved_count}.jpg")
            cv2.imwrite(filename, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"✅ Done! Saved {saved_count} frames to: {output_folder}\n")

# 🏁 Main Execution
if __name__ == "__main__":
    # Automatically get the path to your Downloads folder
    downloads_path = str(Path.home() / "Downloads" / "extracted_frames")

    # List of all your video paths
    video_paths = [
        r"C:\Users\abhir\Downloads\VID_20250420_173033.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_070327.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_195051.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_175721.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_070225.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_171022.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_070605.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_073940.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_074307.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_194940.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_070858.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_165342.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_172753.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_071433.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_183631.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_170623.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_183300.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_165119.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_074501.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_170456.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_165204.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_183338.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_173918.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_172931.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_183458.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_165933.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_074427.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_170508.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_180357.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_195308.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_173536.mp4",
        r"C:\Users\abhir\Downloads\VID_20250420_070639.mp4",
        r"C:\Users\abhir\Downloads\VID_20250419_165715.mp4"
    ]

    interval = 125  # Change this to extract more or fewer frames

    for video_path in video_paths:
        extract_frames(video_path, downloads_path, interval)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from collections import Counter
import pandas as pd

# Ensure plots show in the notebook
%matplotlib inline

# Load the trained YOLOv8 model
model = YOLO(r"C:\Users\abhir\yolov5\runs\detect\crowd_yolo\weights\best.pt")  # Path to your trained model

# Function to process video and estimate crowd density
def estimate_crowd_density(video_path, max_display_frames=5):
    # Initialize video capture
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise ValueError(f"Error opening video file: {video_path}")

    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Lists to store results
    people_counts = []
    timestamps = []
    density_categories = []
    annotated_frames = []

    frame_idx = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLOv8 inference
        results = model(frame, conf=0.5, iou=0.7)  # Adjust confidence and IoU thresholds as needed

        # Count people (assuming class 0 is 'person')
        person_count = sum(1 for box in results[0].boxes if int(box.cls) == 0)
        people_counts.append(person_count)

        # Calculate timestamp
        timestamp = frame_idx / fps
        timestamps.append(timestamp)

        # Classify density
        if person_count <= 10:
            density = "Low"
        elif person_count <= 25:
            density = "Medium"
        else:
            density = "High"
        density_categories.append(density)

        # Draw bounding boxes and density on frame
        annotated_frame = results[0].plot()
        cv2.putText(
            annotated_frame,
            f"Density: {density} ({person_count} people)",
            (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2,
        )

        # Store annotated frame for display (up to max_display_frames)
        if len(annotated_frames) < max_display_frames:
            annotated_frames.append(annotated_frame)

        frame_idx += 1

    cap.release()

    # Create DataFrame for analysis
    df = pd.DataFrame({
        "Timestamp": timestamps,
        "People_Count": people_counts,
        "Density_Category": density_categories
    })

    return df, annotated_frames

# Function to display results and plots
def display_results(df, annotated_frames):
    # Display annotated frames
    num_frames = len(annotated_frames)
    if num_frames > 0:
        plt.figure(figsize=(15, 5 * num_frames))
        for i, frame in enumerate(annotated_frames):
            plt.subplot(num_frames, 1, i + 1)
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.title(f"Frame {i + 1} with Annotations")
            plt.axis('off')
        plt.tight_layout()
        plt.show()

    # Plot 1: Time vs People Count
    plt.figure(figsize=(10, 6))
    plt.plot(df["Timestamp"], df["People_Count"], 'b-', label="People Count")
    plt.title("Time vs People Count")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Number of People")
    plt.grid(True)
    plt.legend()
    plt.show()

    # Plot 2: Density Category Distribution
    density_counts = Counter(df["Density_Category"])
    categories = ["Low", "Medium", "High"]
    counts = [density_counts.get(cat, 0) for cat in categories]

    plt.figure(figsize=(8, 6))
    plt.bar(categories, counts, color=['green', 'orange', 'red'])
    plt.title("Density Category Distribution")
    plt.xlabel("Density Category")
    plt.ylabel("Number of Frames")
    plt.grid(True, axis='y')
    plt.show()

    # Display DataFrame summary
    print("Results DataFrame (first 5 rows):")
    print(df.head())
    print("\nSummary Statistics:")
    print(df.describe())

# Main execution
if __name__ == "__main__":
    # Specify your test video path
    test_video_path = r"C:\Users\abhir\Downloads\Untitled video - Made with Clipchamp.mp4"  # Update this path

    # Estimate crowd density
    results_df, annotated_frames = estimate_crowd_density(test_video_path)

    # Display results
    display_results(results_df, annotated_frames)
```
